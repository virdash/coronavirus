```python
search = tweepy.Cursor(api.search, q="#COVID-19" + " -filter:retweets", result_type="recent", lang="en").items(100)
```

```python
from datetime import date, time
```

```python
tweets_list = []
for item in search:
    tweets_list.append({"text":item.text, "date": (item.created_at).date(), "time": (item.created_at).time(), "handle": item.user.screen_name, "url": "https://twitter.com/" + item.user.screen_name + "/status/" + str(item.id) })
```
